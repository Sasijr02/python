```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)

g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
